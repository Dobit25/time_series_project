# Check stationary

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import os

# --- Configuration ---
data_dir = r"D:\Downloads\Time Series\TIME SERIES FINAL\Cleaned"

# List of files from your image
files_to_process = [
    "EldenRing_Cleaned.csv",
    "HuntShowdown_Cleaned.csv",
    "RDR2_Cleaned.csv",
    "ReadyorNot_Cleaned.csv"
]

# Helper function for Stationarity Testing
def test_stationarity(series, title):
    series = series.dropna()
    print(f"--- Stationarity Tests for: {title} ---")
    
    # ADF Test
    adf_test = adfuller(series, autolag='AIC')
    adf_stat, adf_pval = adf_test[0], adf_test[1]
    adf_conc = "Stationary" if adf_pval < 0.05 else "Non-Stationary"
    
    # KPSS Test
    kpss_test = kpss(series, regression='c', nlags='auto')
    kpss_stat, kpss_pval = kpss_test[0], kpss_test[1]
    kpss_conc = "Non-Stationary" if kpss_pval < 0.05 else "Stationary"
    
    print(f"  ADF Test  | Stat = {adf_stat:>8.4f} | p-value = {adf_pval:.4f} => [ {adf_conc} ]")
    print(f"  KPSS Test | Stat = {kpss_stat:>8.4f} | p-value = {kpss_pval:.4f} => [ {kpss_conc} ]\n")

# =====================================================================
# Main Loop
# =====================================================================
for file_name in files_to_process:
    file_path = os.path.join(data_dir, file_name)
    game_name = file_name.split("_")[0]
    
    print(f"\n{'='*60}")
    print(f"PROCESSING: {game_name}")
    print(f"{'='*60}\n")

    # Load Data
    df = pd.read_csv(file_path, parse_dates=['Date'])
    df.set_index('Date', inplace=True)
    df.sort_index(inplace=True)

    # 1. Log Transformation
    df['Log_Player'] = np.log(df['Avg_Player'] + 1)

    # 2. Time Series Plotting
    plt.figure(figsize=(15, 6))
    plt.plot(df.index, df['Log_Player'], label='Log(Avg_Player)', color='teal', linewidth=1.5)

    in_sale = False
    start_sale_date = None

    for date, row in df.iterrows():
        if row['Is_Sale'] == 1 and not in_sale:
            start_sale_date = date
            in_sale = True
        elif row['Is_Sale'] == 0 and in_sale:
            plt.axvspan(start_sale_date, date, color='red', alpha=0.2)
            in_sale = False

    if in_sale:
        plt.axvspan(start_sale_date, df.index[-1], color='red', alpha=0.2, label='Sale Period')
    else:
        plt.axvspan(df.index[0], df.index[0], color='red', alpha=0.2, label='Sale Period', visible=False)

    plt.title(f'Log_Player over Time ({game_name})')
    plt.xlabel('Date')
    plt.ylabel('Log(Avg_Player + 1)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # 3. Autocorrelation Analysis
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))
    plot_acf(df['Log_Player'].dropna(), lags=35, ax=axes[0], title=f'ACF - {game_name}')
    plot_pacf(df['Log_Player'].dropna(), lags=35, ax=axes[1], title=f'PACF - {game_name}')
    plt.suptitle(f'Autocorrelation Analysis for {game_name}', fontsize=14)
    plt.tight_layout()
    plt.show()

    # 4. Stationarity Testing
    test_stationarity(df['Log_Player'], f"{game_name} Log_Player (Level)")
    test_stationarity(df['Log_Player'].diff(), f"{game_name} Log_Player (1st Diff)")

    # 5. OLS Residual Stationarity
    X_cols = ['Days_Since_Release', 'Discount_%', 'Is_Weekend', 'Is_Holiday']
    # Check if all columns exist in the dataframe to avoid errors
    available_cols = [c for c in X_cols if c in df.columns]
    
    X = df[available_cols].copy()
    X = sm.add_constant(X)
    y = df['Log_Player']

    ols_model = sm.OLS(y, X, missing='drop').fit()
    residuals = ols_model.resid

    print(f"--- OLS Regression Summary ({game_name}) ---")
    print(ols_model.summary().tables[1])
    print("\n")

    test_stationarity(residuals, f"{game_name} OLS Residuals")
    
    # Close plots to free up memory before next iteration
    plt.close('all')
```

# Build model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
import os
import warnings

# Tắt cảnh báo để output gọn gàng
warnings.filterwarnings("ignore")

# --- 1. Data Preparation Function ---
def prepare_data_scaled(file_path):
    df = pd.read_csv(file_path, parse_dates=['Date'])
    df.set_index('Date', inplace=True)
    df.sort_index(inplace=True)
    
    # Ép tần suất Daily để ARIMA hiểu khoảng cách thời gian
    df = df.asfreq('D')
    
    # Target Variable: Log transform để giảm phương sai
    df['Log_Player'] = np.log(df['Avg_Player'] + 1)
    
    # Điền giá trị trống phát sinh do asfreq (nếu có)
    df['Log_Player'] = df['Log_Player'].interpolate(method='linear')
    
    # Feature Scaling: Giúp MLE hội tụ nhanh hơn
    df['Discount_Ratio'] = df['Discount_%'] / 100.0
    df['Years_Since_Release'] = df['Days_Since_Release'] / 365.0
    
    # Fill các biến ngoại sinh bị trống sau khi asfreq
    cols_to_fill = ['Years_Since_Release', 'Discount_Ratio', 'Is_Weekend', 'Is_Holiday']
    df[cols_to_fill] = df[cols_to_fill].ffill().bfill()
    
    endog = df['Log_Player']
    exog = df[['Years_Since_Release', 'Discount_Ratio', 'Is_Weekend', 'Is_Holiday']]
    
    return df, endog, exog

data_dir = r"D:\Downloads\Time Series\TIME SERIES FINAL\Cleaned"
games_groups = {
    "Group A (Multiplayer / Instant Demand)": {
        "Hunt Showdown": "HuntShowdown_Cleaned.csv",
        "Ready or Not": "ReadyorNot_Cleaned.csv"
    },
    "Group B (Single-player / Stockpiling)": {
        "Elden Ring": "EldenRing_Cleaned.csv",
        "Red Dead Redemption 2": "RDR2_Cleaned.csv"
    }
}

# --- 2. Fit Strict ARMAX(1,0,0) Models ---
for group_name, games in games_groups.items():
    print("=" * 80)
    print(f"SCALED ARMAX(1,0,0) FOR {group_name.upper()}")
    print("=" * 80)
    
    for game_name, file_name in games.items():
        print(f"\n--- Model for: {game_name} ({file_name}) ---")
        file_path = os.path.join(data_dir, file_name)
        
        try:
            df, y, X = prepare_data_scaled(file_path)
            
            print("Fitting ARMAX(1,0,0) using L-BFGS optimizer...")
            
            model_spec = ARIMA(
                endog=y, 
                exog=X,
                order=(1, 0, 0),
                enforce_stationarity=True,
                enforce_invertibility=True
            )
            
            # SỬA LỖI TẠI ĐÂY: Truyền maxiter vào method_kwargs
            final_model = model_spec.fit(
                method='statespace',
                method_kwargs={'maxiter': 500} # Tăng mạnh vòng lặp để ép hội tụ
            )
            
            # --- 3. Output Requirements ---
            try:
                discount_coef = final_model.params['Discount_Ratio']
                discount_pval = final_model.pvalues['Discount_Ratio']
                print("\n>> ECONOMIC INTERPRETATION <<")
                print(f"Price Elasticity (Discount_Ratio Coef): {discount_coef:.5f} (p-value: {discount_pval:.4f})")
                if discount_pval < 0.05 and discount_coef > 0:
                    print(" -> Success: Discount has a significant POSITIVE impact on players.")
                elif discount_pval < 0.05:
                    print(" -> Discount is significant, but unexpectedly negative.")
                else:
                    print(" -> Discounts do not show a statistically significant relationship at the 5% level.")
            except KeyError:
                print("Discount_Ratio was not found in the parameters.")

            # Residual Diagnostics
            print("\n>> RESIDUAL DIAGNOSTICS (Ljung-Box Test) <<")
            residuals = final_model.resid
            lb_test = acorr_ljungbox(residuals, lags=10, return_df=True)
            print("P-values for Lags 1-10 (ideal is > 0.05):")
            print(lb_test['lb_pvalue'].to_frame().T)
            
            # Visualizing
            df['Fitted'] = final_model.fittedvalues
            df_last_year = df.tail(365)
            
            plt.figure(figsize=(14, 5), dpi=100)
            plt.plot(df_last_year.index, df_last_year['Log_Player'], label='Actual Log_Player', color='black', alpha=0.6)
            plt.plot(df_last_year.index, df_last_year['Fitted'], label='Fitted ARMAX(1,0,0)', color='darkorange', linewidth=2)
            plt.title(f'{game_name} - ARMAX Analysis', fontsize=12, fontweight='bold')
            plt.ylabel('Log(Avg_Player)')
            
            # Green shade for sale days
            sale_mask = df_last_year['Discount_Ratio'] > 0
            plt.fill_between(df_last_year.index, df_last_year['Log_Player'].min(), df_last_year['Log_Player'].max(), 
                             where=sale_mask, color='green', alpha=0.1, label='Active Sale')

            plt.legend(loc='lower left')
            plt.grid(True, linestyle=':', alpha=0.5)
            plt.tight_layout()
            plt.show()
            
        except Exception as e:
            print(f"Model for {game_name} failed: {e}")
```
